This notebook recreates the main exploratory data analysis (EDA) graphs using a custom pharma-inspired color palette, with the goal of exporting professional visuals for the project presentation.


```python
import pandas as pd

df = pd.read_csv("data_sample_full_named.csv")
adr_columns = df.columns[1:25]
df["total_adr"] = df[adr_columns].sum(axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pharma-inspired color palette
main_color = "#1f77b4"   # light blue
secondary_color = "#2ca02c"  # clinical green
text_color = "#7f7f7f"
title_color = "#0d3b66"
sns.set_style("whitegrid")
```

## ADR Frequency

```python
adr_counts = df[adr_columns].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=adr_counts.index, y=adr_counts.values, color=main_color)
plt.xticks(rotation=90, color=text_color)
plt.title("Frequency of Each Adverse Drug Reaction", color=title_color, fontsize=14)
plt.ylabel("Number of Compounds", color=text_color)
plt.tight_layout()
plt.savefig("pharma_adr_frequency.png", dpi=300)
plt.show()
```

## Number of ADRs per Compound

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["total_adr"], bins=20, kde=False, color=secondary_color)
plt.title("Distribution of ADR Count per Compound", color=title_color)
plt.xlabel("ADR Count", color=text_color)
plt.ylabel("Number of Compounds", color=text_color)
plt.tight_layout()
plt.savefig("pharma_adr_per_compound.png", dpi=300)
plt.show()
```

## ADR Correlation Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[adr_columns].corr(), cmap="BuGn", annot=False)
plt.title("Correlation Between ADRs", color=title_color)
plt.tight_layout()
plt.savefig("src/img/pharma_adr_correlation_heatmap.png", dpi=300)
plt.show()
```
